# Looping Statements

```python
# print the number form 1 to 100
for i in range(1,101):
    print(i)
```

```python
# print the number form 50 to 0 with a gap of 5
for i in range(50, -1, -5):
    print(i)
```

```python
# itreating through string as well
name = "amit kumar singh"
for char in name:
    print(char, end=" ")
```

```python
# By making the use of while loop we need to print the numbers from o to 50
i = 0
while i<=50:
    print(i)
    i += 1
    
```

```python
for i in range(1,50):
    if(i%2==0):
        print("Even", i)
```

```python
i = 0
while i<=50:
    if(i%2==0):
        print("Even", i)
    i += 1
```

```python
for i in range(1,50):
    if(i==18):
        break
    print(i)
```

```python
for i in range(51):
    if(i%2==0):
        print(i)
    else:
        continue
```

```python
for i in range(21):
    if(i==17):
        continue
    print(i)
```

## Nested Loops

```python
for i in range(6):
    for j in range(4):
        print(i, j)
```

```python
## printing the sum of first N natural numbers where Natural Numbers ----> 1,2,3....n

n_value = int(input("Enter the value of n: "))
sum = 0
for i in range(1, n_value+1):
    sum += i
print("The sum of first", n_value, "natural numbers is:", sum)
```
