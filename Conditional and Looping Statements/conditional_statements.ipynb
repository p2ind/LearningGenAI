# Conditional Statements

```python
amit_height = int(input("Enter what is your height is amit"))
shivy_height = int(input("Enter what is your height is shivy"))

if amit_height > shivy_height:
    print("Amit is taller than Shivy")
else:
    print("Shivy is taller than Amit")
```

```python
year = int(input("Enter the year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```
