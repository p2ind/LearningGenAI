# Task 2: Categories (Sets)

```python
# Product categories list
categories = [
    "Electronics",
    "Accessories",
    "Accessories",
    "Electronics",
    "Office",
    "Audio",
    "Electronics",
    "Audio"
]
```

```python
#1. Create set
categories_set = set(categories)
```

```python
print("Unique Categories:")
print(categories_set)
```

```python
#2 Add new category
categories_set.add("Gaming")
# Add duplicate category
categories_set.add("Audio")
```

```python
print("\nAfter Adding Categories:")
print(categories_set)
```

```python
#3. Check category exists
print("\nIs 'Office' present?")
print("Office" in categories_set)
```

```python
# Extra (Optional)
print("\nTotal Unique Categories:")
print(len(categories_set))
```
