# Task 4: Combined Operations

```python
# Products list
products = [
    "Laptop",
    "Mouse",
    "Keyboard",
    "Monitor",
    "Printer",
    "Speaker"
]

# Categories list
categories = [
    "Electronics",
    "Accessories",
    "Accessories",
    "Electronics",
    "Office",
    "Audio"
]

# Price dictionary
price_dict = {
    "Laptop": 55000,
    "Mouse": 900,
    "Keyboard": 1500,
    "Monitor": 12000,
    "Printer": 8000,
    "Speaker": 3000
}
```

```python
# Create catalog list
catalog = []

for i in range(len(products)):
    product_name = products[i]
    price = price_dict[product_name]
    category = categories[i]

    catalog.append((product_name, price, category))

print("Catalog:")
print(catalog)
```

```python
# Create category_to_products dictionary
category_to_products = {}

for item in catalog:
    product_name = item[0]
    category = item[2]

    if category not in category_to_products:
        category_to_products[category] = []

    category_to_products[category].append(product_name)

print("\nCategory to Products:")
print(category_to_products)
```

```python
# Find category with maximum products
max_category = max(
    category_to_products,
    key=lambda x: len(category_to_products[x])
)

print("\nCategory with Maximum Products:")
print(max_category)

print("\nProducts in that Category:")
print(category_to_products[max_category])
```
