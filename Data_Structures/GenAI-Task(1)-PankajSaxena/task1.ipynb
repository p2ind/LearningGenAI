# Task 1: Product Collections (Lists & Tuples)

```python
# 1. Create a product_list at least contain 6 product names (string).
product_list = ["Laptop", "Smartphone", "Headphones", "Camera", "Smartwatch", "Tablet"]

# 2. Create a tuple named sample_product that stores (product_name, price, category) for one product.
sample_product = ("Laptop", 999.99, "Electronics")
```

```python
# 3. Print the 2nd and last product from the product list.
print("2nd product:", product_list[1])
print("Last product:", product_list[-1])
```

```python
# 4. Append two new product names to products and then print the updated list.
product_list.append("Monitor")
product_list.append("Keyboard")
print("Updated product list:", product_list)
```

```python
# Extra(optinal): Convert sample_product into a list, change its price, and convert it back to a tuple.
sample_product_list = list(sample_product)
sample_product_list[1] = 899.99 # Change the price
updated_sample_product = tuple(sample_product_list)
print("Updated sample product:", updated_sample_product)
```
