# Task 3: Product Pricing (Dictionaries)

```python
#1. Create a dicitonary price_dict where keys are product names and values are prices (integers or floats). Include at least 6 entries.
price_dict = {
    "Laptop": 999.99,
    "Smartphone": 499.99,
    "Headphones": 199.99,
    "Monitor": 299.99,
    "Keyboard": 89.99,
    "Mouse": 49.99
}
```

```python
print("Original Dictionary:")
print(price_dict)
```

## 2. Write small code blocks to:

```python
# Add a new product with prict to prict_dict
price_dict["Tablet"] = 299.99
print(f"New product 'Tablet' added successfully.")
print(price_dict)
```

```python
# Update the price of an existing product
price_dict["Laptop"] = 899.99
print(f"Product 'Laptop' price updated successfully.")
print(price_dict)
```

```python
# Remove product safely
product_to_remove = "Tablet"
if product_to_remove in price_dict:
    price_dict.pop(product_to_remove, None)
    print(f"Product '{product_to_remove}' removed successfully.")
else:
    print(f"Product '{product_to_remove}' not found.")

print("\nUpdated Dictionary:")
print(price_dict)
```

## 3. Print the average price of all products (use only dictionary operations and basic arithmetic).

```python
#3. Print the average price of all products (use only dictionary operations and basic arithmetic).
total_price = sum(price_dict.values())
number_of_products = len(price_dict)
average_price = total_price / number_of_products
print(f"\nAverage price of all products: ${average_price:.2f}")
```

```python
#4. Extra (optinal): Print the product with both the maximum and minimum prices.
max_price_product = max(price_dict, key=price_dict.get)
min_price_product = min(price_dict, key=price_dict.get)
print(f"\nProduct with the maximum price: {max_price_product} (${price_dict[max_price_product]})")
print(f"Product with the minimum price: {min_price_product} (${price_dict[min_price_product]})")
```
