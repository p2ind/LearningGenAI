# Tuples

```python
# Tuples example
my_tuple = (1, 2, 3, 4, 5)
print(my_tuple)
print(type(my_tuple))
```

```python
# tuples are immutable, meaning you cannot change their contents after they are created. However, you can perform various operations on tuples, such as indexing, slicing, and concatenation.
# tuples are also capable of storing the data of any data type
# Example of a tuple containing different data types
mixed_tuple = (1, "Hello", 3.14, [1, 2, 3], (4, 5))
print(mixed_tuple)
# You can also use tuples to unpack values into variables
a, b, c, d, e = mixed_tuple
print(a)  # Output: 1
print(b)  # Output: Hello
print(c)  # Output: 3.14
print(d)  # Output: [1, 2, 3]
print(e)  # Output: (4, 5)
# Tuples can also be used as keys in dictionaries, while lists cannot because they are mutable.
my_dict = {my_tuple: "This is a tuple key"}
print(my_dict)
# You can also create a tuple with a single element by including a comma after the element
single_element_tuple = (42,)
print(single_element_tuple)
print(type(single_element_tuple))
```

```python
mixed_tuple[::-1]
```

## Packing and unpacking of Tuples

```python
# Packing of tuples
packed_tuple = 1, "Hello", 3.14
print(packed_tuple)
type(packed_tuple)
```

```python
# unpacking of tuples
a,b,c,d,e = (1,2,3,4,5)
```

```python
a
```

```python
# unpacking with a *
first_value, *middle_values, last_value = (1,2,3,4,5)
print(first_value)  # Output: 1
print(middle_values)  # Output: [2, 3, 4]
print(last_value)  # Output: 5
```

## Nested List

```python
Nested_List = [[1,2,3,16], [4,5,6], [11,14,17], [12,13,18,15,19]]
print(Nested_List[0][:])
```

```python
print(Nested_List[0][2])
```

```python
Nested_List[3][:]
```

```python
Nested_List
```

```python
# Iterating over a Tuple inside a Tuple
for i in Nested_List:
    for j in i:
        print(j, sep=" ")
    print(" ")
```
