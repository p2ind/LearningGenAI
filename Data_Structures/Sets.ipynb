```python
## Methods in sets
set_value.add(6) # add an element to the set
print("Added", set_value)
set_value.remove(3) # remove an element from the set
print("Removed 3:", set_value)
set_value.discard(1) # remove an element from the set if it exists, otherwise do nothing
print("Discarded 10:", set_value)   
set_value.clear() # remove all elements from the set
print("Cleared:", set_value)
```
