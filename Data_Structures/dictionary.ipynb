# Dictionary Methods

## Keys()
## values()
## items()
## get()

```python
dictionary = {"Name": "John","Fruit":"Apple","Vegetable":"Carrot", "List_items":[1,2,3,4], "Numbers": 34}
print(dictionary)
```

```python
# Keys() --> Returns a view object that displays a list of all the keys in the dictionary.
print(dictionary.keys())
```

```python
# values() --> Returns a view object that displays a list of all the values in the dictionary.
print(dictionary.values())
```

```python
# items() --> Returns a view object that displays a list of dictionary's key-value tuple pairs.
print(dictionary.items())
```

```python
#get() --> Returns the value of the specified key. If the key does not exist, it returns a default value (None if not specified).
print(dictionary.get("Name"))  # Output: John
print(dictionary.get("NonExistentKey", "Default Value"))  # Output: Default Value
```

```python
# Shallow Copy --> Creates a new dictionary that references the same objects as the original dictionary. Changes to mutable objects in the original dictionary will affect the shallow copy.
shallow_copy = dictionary.copy()
shallow_copy["Name"] = "Jane"
print("Original Dictionary:", dictionary)
print("Shallow Copy:", shallow_copy)
# Deep Copy --> Creates a new dictionary that is a completely independent copy of the original dictionary. Changes to mutable objects in the original dictionary will not affect the deep copy.
import copy
deep_copy = copy.deepcopy(dictionary)
deep_copy["Name"] = "Alice"
print("Original Dictionary:", dictionary)
print("Deep Copy:", deep_copy)
    
```

## Nested Dictionary

```python
# I want to store the information for multiple students like there name, marks, grade
students = {
    "Student1": {"Name": "John", "Marks": 85, "Grade": "A"},
    "Student2": {"Name": "Alice", "Marks": 78, "Grade": "B"},
    "Student3": {"Name": "Bob", "Marks": 92, "Grade": "A+"}
}
print(students)
```

```python
#iterate through the dictionary
for student, details in students.items():
    print(f"{student}: {details}")
    for key, value in details.items():
        print(f"  {key}: {value}")
```

## Dictionary Comprehension

## Dictionary comprehension helps us with using Looping Statement along with conditional statement inside our Dictionary

```python
comprehensive = {x**2 for x in range(5)}
```

```python
comprehensive
```

```python
comprehensive = {x:x**2 for x in range(11)}
```

```python
comprehensive
```

```python
even_dict = {x:x**2 for x in range(1,11) if x%2==0}
even_dict
```

```python
## Merged Dictionary    
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}
merged_dict = {**dict1, **dict2}
print(merged_dict)
```

## Additional Functions in a List are:
## sum()
## max()
## min()
## len()

```python
lists = {x for x in range(17)}
lists
```

```python
#sum() --> Returns the sum of all the items in an iterable (like a list or set) or the sum of all the values in a lists.
sum_of_values = sum(lists)
print(sum_of_values)
```

```python
# max() --> Returns the largest item in an iterable or the largest of two or more arguments.
max_value = max(lists)
print(max_value)
```
